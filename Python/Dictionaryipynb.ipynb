```python
dict = {}
dict[0] = {"Book": "SN Dey", "Page": 18}
dict[1] = {"Book": "RD Sharma", "Page": 29}
dict[2] = {"Book": "Dwari Maity", "Page": 88}
dict[3] = {"Book": "Thareja", "Page": 100}

for i,j in dict.items():
  print(i,"-",j)

for i in dict:
    print(i,"->",dict[i])

for key, value in dict.items():
  print(key,"-",value)

```
