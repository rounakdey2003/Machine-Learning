```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/csv/Dataset/Indian House Prices.csv')
df
```

```python
X = df['Area'].loc[:20]
y = df['Price'].loc[:20]

import matplotlib.pyplot as plt
plt.scatter(X,y,marker='o',c='r')
plt.title('House Prediction')
plt.xlabel('Area of the House')
plt.ylabel('Price(Rs)')
plt.show()
```

```python
w=0.05
b=0.10

import numpy as np

def cost_function(x,w,b):
  m = x.shape[0]
  f_wb = np.zeros(m)
  for i in range(m):
    f_wb[i] = w*x[i]+b
  return f_wb
```

```python
call_cost_Function = cost_function(X,w,b)
plt.plot(X,call_cost_Function,c='b',label='Prediction')
plt.scatter(X,y,marker='o',c='r',label='Actual')
plt.title('House Prediction')
plt.xlabel('Area of the House')
plt.ylabel('Price(Rs)')
plt.show()
```
