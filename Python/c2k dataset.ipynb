```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('c2k_data_comma.csv')
df.head()

nr = df['nr']
i1_legid = df['i1_legid']
i1_rcs_p = df['i1_rcs_p']

figure,axs=plt.subplots(3)
axs[0].plot(nr,color='orange')
axs[1].plot(i1_legid,color='blue')
axs[2].plot(i1_rcs_p,color='red')
plt.show()
```
