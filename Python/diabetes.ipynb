```python
import numpy as np
from sklearn.datasets import load_diabetes
import pandas as pd

data = load_diabetes()
data
```

```python
df = pd.DataFrame(data.data,index=data.target,columns=data.feature_names)
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plotnumber=1
for column in df:
    if plotnumber<=10:
        plt.subplot(3,4,plotnumber)
        sns.distplot(df[column],color='cyan')
        plt.xlabel(column,fontsize=15)
    plotnumber+=1
plt.tight_layout()
plt.show()
```

```python
plotnumber=1
for column in df:
    if plotnumber<=10:
        plt.subplot(3,4,plotnumber)
        sns.scatterplot(x=df['age'],y=df[column],color='orange')
    plotnumber+=1
plt.tight_layout()
plt.show()
```
